```python
import json
import pandas as pd
```

```python
with open('ContentVisio.json') as json_file:
    data = json.load(json_file)
```

```python
def extract_values(obj, key):
    """Pull all values of specified key from nested JSON."""
    arr = []

    def extract(obj, arr, key):
        """Recursively search for values of key in JSON tree."""
        if isinstance(obj, dict):
            for k, v in obj.items():
                if isinstance(v, (dict, list)):
                    extract(v, arr, key)
                elif k == key:
                    arr.append(v)
        elif isinstance(obj, list):
            for item in obj:
                extract(item, arr, key)
        return arr

    results = extract(obj, arr, key)
    return results
```

```python
def workflowStrip (data):
    workflowdict={}
    counter = 1
    dictcount = 1
    for element in data:
        if counter%2 != 0:
            workflowdict[dictcount] = {}
            workflowdict[dictcount]['Actor'] = element
        else:
            workflowdict[dictcount]['Action'] = element
            dictcount=dictcount+1
        counter = counter+1
    return workflowdict
```
